```r
projFoxn1ko = loadArchRProject("ML_0409")
```

```r
projFoxn1ko
```

```r
projFoxn1ko <- filterDoublets(projFoxn1ko)
```

```r
projFoxn1ko <- addIterativeLSI(
    ArchRProj = projFoxn1ko,
    useMatrix = "TileMatrix", 
    name = "IterativeLSI_test", 
    iterations = 2, 
    clusterParams = list( #See Seurat::FindClusters
        resolution = c(2), 
        maxClusters=12,
        sampleCells = nCells(projFoxn1ko), 
        n.start = 10
    ), 
    sampleCellsPre = NULL,
    sampleCellsFinal = NULL,
    varFeatures = 20000, 
    dimsToUse = 1:35,
    projectCellsPre = TRUE, force=TRUE
    # sampleCells = FALSE
)

# IterativeLSI_6_20k_30

projFoxn1ko <- addClusters(
    input = projFoxn1ko,
    reducedDims = "IterativeLSI_test",
    method = "Seurat",
    name = "Clusters_test_1.6",
    resolution = 1.6, force=TRUE, maxClusters=100
)


projFoxn1ko <- addUMAP(
    ArchRProj = projFoxn1ko, 
    reducedDims = "IterativeLSI_test", 
    name = "UMAP_test",
    # scaleDims = FALSE,
    #dimsToUse = 2:15,
    nNeighbors = 30,
    minDist = 0.2, 
    metric = "cosine",force=TRUE
)

```

```r
p1 <- plotEmbedding(ArchRProj = projFoxn1ko, size = 1.0,colorBy = "cellColData",name = "Sample", embedding = "UMAP_test")

p2 <- plotEmbedding(ArchRProj = projFoxn1ko, size = 1.0,colorBy = "cellColData", name = "nFrags", embedding = "UMAP_test")

p3 <- plotEmbedding(ArchRProj = projFoxn1ko, size = 1.0,colorBy = "cellColData", name = "DoubletScore", embedding = "UMAP_test")


p5 <- plotEmbedding(ArchRProj = projFoxn1ko, size = 1.0,colorBy = "cellColData", name = "Clusters_test_1.6", embedding = "UMAP_test")


plotPDF(p1,p2,p3,p5, name = "FigS3a.pdf", ArchRProj = projFoxn1ko, addDOC = FALSE, width = 5, height = 5)



```
