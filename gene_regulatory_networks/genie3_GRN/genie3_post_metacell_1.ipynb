```r
PERMITTED_DGE_TYPES = list(
  dge_type = c("metacell", "full", "velocyto_loom", 
               "velocyto_spliced",   "velocyto_spliced_metacell", 
               "velocyto_unspliced", "velocyto_unspliced_metacell"),
  corresponding_metadata_type = c("metacell", "full", "full", 
                                  "full", "metacell", 
                                  "full", "metacell"),
  metacell_precursors     = c("full",     "velocyto_spliced",          "velocyto_unspliced"), 
  corresponding_metacells = c("metacell", "velocyto_spliced_metacell", "velocyto_unspliced_metacell")
)
```

```r
get_metadata_type = function(dge_type) {
  # assertthat::assert_that(dge_type %in% PERMITTED_DGE_TYPES$dge_type)
  setNames( PERMITTED_DGE_TYPES$corresponding_metadata_type, PERMITTED_DGE_TYPES$dge_type)[[dge_type]]
}
get_metacell_type = function(dge_type){
  # assertthat::assert_that(dge_type %in% PERMITTED_DGE_TYPES$metacell_precursors)
  setNames( PERMITTED_DGE_TYPES$corresponding_metacells, PERMITTED_DGE_TYPES$metacell_precursors)[[dge_type]]
}
```

```r
#' Load cell metadata
#' 
#' @param type full data or metacell?
#' 
#' Returns a dataframe
load_cell_metadata = function( dataset_name = NULL, type = "full", stop_for_missing_file = T ){
  
    # assertthat::assert_that(type %in% PERMITTED_DGE_TYPES$dge_type)
  
  file_path = here("data", dataset_name, paste0(get_metadata_type(type), "_metadata.csv"))
  print(file_path)
  if(!file.exists(file_path) ){
    if(stop_for_missing_file){
      stop(": Metadata not available.\n" %>% paste0(dataset_name, .))
    } else {
      return()
    }
  }
  metadata_df = read.csv(file_path, stringsAsFactors = F)
  rownames(metadata_df) = metadata_df[["CELL_ID"]]
  metadata_df
}
```

```r
#' Apply a function to all dge's from a given source. 
#' 
#' @param dataset_name Which data source; e.g. "pharynx_atlas"
#' @param test If true, the just run on a randomly selected DGE. If false, run on all of them.
#' @param type "full" for cells, "metacell" for metacells
#' @param omit Skip these.
#' @param group_by If you want to group by a different variable, for example "BATCH". This 
#' may become more flexible in the future, but for now you can only group by a variable
#' in which DGE is nested (a strictly "coarser" variable). 
#' 
iterate_within_source = function(dataset_name, FUN, 
                                 test = F,
                                 omit = NULL, 
                                 group_by = "DGE", 
                                 ...){
  X = load_cell_metadata(dataset_name) 
  X = X[!(X[[group_by]] %in% omit), ]
  available_dges = X %>% extract(c(group_by, "DGE") ) %>%
    table %>% as.data.frame %>% subset(Freq > 0) %>% 
    (plyr::dlply)(.variables = group_by, .fun = extract2, "DGE") %>% 
    lapply(as.character)
  
  if(length(available_dges)==0){
    warning(paste0("No expression data available: ", dataset_name, " type = ", type))
    return()
  }
  if(test){
    available_dges = sample(available_dges, 1)
  }
  ma = list(...)
  print(available_dges)
  mapply(FUN, dataset_name, available_dges, MoreArgs = ma, SIMPLIFY = F) %>% setNames(available_dges)
}
```

```r
#' Load expression data from a given dge matrix.
#' 
#' @param dataset_name A subfolder of "data/".
#' @param dge_name A dge in that subfolder. Can come from the DGE field of the cell metadata.
#' @param type Try "full", "metacell", or anything in PERMITTED_DGE_TYPES$dge_type .
#' @param fix_cellnames Cell ID's get passed through this function. 
#' @param mode If simple, return what you see in the file. 
#' If helpful, look for cells missing from metadata. If any found, throw a warning. They are assumed to be empties or low quality and they are not returned.
#' If strict, look for cells missing from metadata. If any found, throw an error.
#' Returns a Matrix
load_expr_by_dge = function( dataset_name, dge_name, type = "full", mode, fix_cellnames = NULL, ... ){
  assertthat::assert_that(mode %in% c("simple", "helpful", "strict"))
  assertthat::assert_that(type %in% PERMITTED_DGE_TYPES$dge_type)
  
  # Find it
  dge_location = here("data", dataset_name, type %>% gsub('(un)*spliced$', "loom", .), dge_name) 
  print(dge_location)
  if( type %in% c("velocyto_loom", "velocyto_spliced", "velocyto_unspliced")){
    dge_location %<>% paste0(".loom")
  }
  if(is.null(dge_location)){
    stop("Error locating expression matrix! Please report this error to Eric.\n")
  }
  
  # read anything -- loom, .tsv, .csv, 10Xv2, 10Xv3
  X = read_one_dge_flexibly(dge_location, ...)
  
  # deal with velocity-associated bullshit
  if( type == "velocyto_spliced" ){
    X = X[["spliced"]]
  }
  if( type == "velocyto_unspliced" ){
    X = X[["unspliced"]]
  }
  if( !is.null( fix_cellnames )){
    colnames(X) %<>% fix_cellnames
  }
  
  # Return cells depending on mode
  if(mode=="simple"){
    return(X)
  }
  metadata = load_cell_metadata(dataset_name = dataset_name, type = type)
  good_cells = intersect(colnames(X), metadata$CELL_ID)
  if( !identical( good_cells, colnames(X)) ){
    if(mode=="helpful"){
      y = length(colnames(X))
      x = y - length(good_cells)
      warning(
        paste( x, "of", y, 
               "cells missing from metadata! Will omit those from expr matrix.\n",
               dataset_name, "   ", dge_name)
              )
    } 
    if(mode=="strict"){
      check_containment(colnames(X), metadata$CELL_ID, error_message = "Cells missing from metadata!")
      stop("Cells missing from metadata!\n")
    }
  }
  return(X[,good_cells])
}

```

```r
## ------------------------------------------------------------------------
#' Merge a list of digital gene expression matrices.
#'
#' @param dge_list list of matrices with genes as rows and cell barcodes as columns. 
#' @param allow_barcode_overlap Set this to TRUE if matching barcodes in different inputs refer to the same cells, 
#' as in a resequencing of a library. Set this to FALSE (default) if matching barcodes would only occur by coincidence, 
#' as in merging of different replicates for joint analysis.
#' @details `dge_merge_list` converts the digital gene expression matrices to dataframes with genes as columns, merges them, then converts the result back, all without disturbing the gene labels. A gene will be included if it appears in any of the datasets. If a gene appears in one dataset but not another, zeroes will be filled in for missing expression levels.
#' @export
dge_merge_list = function(dge_list, allow_barcode_overlap = F){
  # Allow duplicate cells but not duplicate genes
  all_genes = Reduce( f = union, x = lapply( dge_list, rownames ) )
  all_cells = Reduce( f = c,     x = lapply( dge_list, colnames ) )
  input_dimensions = Reduce(rbind, lapply(dge_list, dim)) %>% matrix(ncol = 2)

  if( allow_barcode_overlap ){
    all_cells = unique( all_cells )
  } else {
    if(anyDuplicated(all_cells)){ 
      warning( "These samples contain overlapping barcodes! 
                \nAttempting to append names(dge_list) to barcodes..." )
      if(is.null(names(dge_list))){ 
        stop("...No names available for dge_list!")
      }
      all_cells = lapply( dge_list, colnames ) %>% 
        mapply( paste0, .,  "|", names(dge_list) ) %>%
        Reduce( f = c,     x = . )
    }
    atat(!anyDuplicated(all_cells)) 
  }
  
  new_dge = Matrix::Matrix( 0, 
                            nrow = length( all_genes ), 
                            ncol = length( all_cells ), 
                            sparse = T ) 
  rownames( new_dge ) = all_genes
  colnames( new_dge ) = all_cells
  for( ii in seq_along( dge_list ) ){
    rr = rownames(dge_list[[ii]])
    cc = colnames(dge_list[[ii]])
    new_dge[rr, cc] = new_dge[rr, cc] + dge_list[[ii]]
    print( paste( "Finished merging dge matrix", names(dge_list)[[ii]] ) )
    dge_list[[ii]] = matrix(); gc()
  }
  #Results should have at least as many genes (barcodes) as the input with the most genes (barcodes) 
  assertthat::assert_that(dim(new_dge)[1] >= max(input_dimensions[,1]))
  assertthat::assert_that(dim(new_dge)[2] >= max(input_dimensions[,2]))
  #If no barcode overlap, should retain all input barcodes.
  if( !allow_barcode_overlap ){
    assertthat::assert_that(dim(new_dge)[2] == sum(input_dimensions[,2]))
  }
  assertthat::assert_that(!is.null(colnames(new_dge)))
  assertthat::assert_that(!is.null(rownames(new_dge)))
  return( new_dge )
}
```

```r
Read10X = function (data.dir = NULL, version = NULL) 
{
  version = ifelse(file.exists(file.path(data.dir, "features.tsv.gz")), 3, 2)
  full.data <- list()
  for(i in seq_along(data.dir)){
    run <- data.dir[i]
    if (!dir.exists(run)) {
      stop("Directory provided does not exist")
    }
    if (!grepl("\\/$", run)) {
      run <- paste(run, "/", sep = "")
    }
    if( version == 3 ){
      barcode.loc <- paste0(run, "barcodes.tsv.gz")
      gene.loc <- paste0(run, "features.tsv.gz")
      matrix.loc <- paste0(run, "matrix.mtx.gz")
      modifier = function(x) gzcon(file(x, open = "r"))
    } else if (version == 2){
      barcode.loc <- paste0(run, "barcodes.tsv")
      gene.loc <- paste0(run, "genes.tsv")
      matrix.loc <- paste0(run, "matrix.mtx")
      modifier = function(x) x
    }
    if (!file.exists(barcode.loc)) {
      stop("Barcode file missing")
    }
    if (!file.exists(gene.loc)) {
      stop("Gene name file missing")
    }
    if (!file.exists(matrix.loc)) {
      stop("Expression matrix file missing")
    }
    matrix.loc = modifier(matrix.loc)
    barcode.loc = modifier(barcode.loc)
    gene.loc = modifier(gene.loc)
    data <- Matrix::readMM(file = matrix.loc)
    cell.names <- readLines(barcode.loc)
    gene.names <- readLines(gene.loc)
    try(close(matrix.loc), silent = T)
    try(close(barcode.loc), silent = T)
    try(close(gene.loc), silent = T)
    
    print(cell.names)
    if (all(grepl(pattern = "\\-1$", x = cell.names))) {
      cell.names <- as.vector(x = as.character(x = sapply(X = cell.names, 
                                                          FUN = ExtractField_me, 
                                                          field = 1,
                                                          delim = "-")))
    }
    # print(cell.names)
    #   print(gene.names)
    rownames(x = data) <- make.unique(names = as.character(x = sapply(X = gene.names,
                                                                      FUN = ExtractField_me, 
                                                                      field = 2,
                                                                      delim = "\\t")))
    if (is.null(x = names(x = data.dir))) {
      if (i < 2) {
        colnames(x = data) <- cell.names
      }
      else {
        colnames(x = data) <- paste0(i, "_", cell.names)
      }
    }
    else {
      colnames(x = data) <- paste0(names(x = data.dir)[i], 
                                   "_", cell.names)
    }
    full.data <- append(x = full.data, values = data)
    full.data <- do.call(cbind, full.data)
    return(full.data)
  }
}
```

```r
ExtractField_me <- function(string, field = 1, delim = "_") {
  fields <- as.numeric(x = unlist(x = strsplit(x = as.character(x = field), split = ",")))
  if (length(x = fields) == 1) {
    return(strsplit(x = string, split = delim)[[1]][field])
  }
  return(paste(strsplit(x = string, split = delim)[[1]][fields], collapse = delim))
}
```

```r
read_one_dge_flexibly = function(to_read, min_umi_count = 100){
  extension = tools::file_ext(gsub(".gz", "", to_read))
  if (extension %in% c("txt", "tsv")){
    X = read.table(to_read, header = T, row.names = 1) %>% as.matrix %>% Matrix(sparse = T)
  } else if (extension=="loom"){
    X = velocyto.R::read.loom.matrices(to_read)
    if( min_umi_count!=100 ){ warning("min_umi_count is ignored when loading velocyto loomfiles.\n") }
    return(X)
  } else if (extension=="csv"){
    X = read.csv(to_read, header = T, row.names = 1, sep = ",", check.names = F) %>% as.matrix %>% Matrix(sparse = T)
  } else {
    assertthat::assert_that(file_test("-d", to_read))
    X = Read10X(to_read, version = 3)
  }
    # print(dim(X))
  X = X[,Matrix::colSums(X) > min_umi_count]
    # print(dim(X))
  X
}
```

```r
#' Save a new gene regulatory (sub-)network.
#' 
#' @param grn_name @param dge_name This function saves to here("networks", grn_name, "networks", subnetwork_name).
#' 
save_grn_by_subnetwork = function( grn_name, subnetwork_name, X, ... ){
  grn_location = here("networks", grn_name, "networks", subnetwork_name)
  cat("Attempting to print relevant metadata. You may need to edit metadata by hand. \n")
  load_grn_metadata() %>% subset(name == grn_name) %>% print
  write.table(X,
              grn_location,
              sep = ",", col.names = F, row.names = F)
  cat("Wrote grn to ", grn_location, "\n")
}
```

```r
#' Load a given gene regulatory (sub-)network.
#' 
#' @param grn_name @param dge_name This function looks in here("networks", grn_name, "networks", subnetwork_name).
#' 
load_grn_by_subnetwork = function( grn_name, subnetwork_name, sort_by_weight = T, ... ){
  grn_location = here("networks", grn_name, "networks", subnetwork_name)
  if(is.null(grn_location)){
    stop("Error locating grn! Please report this error to Eric.\n")
  }
  extension = tools::file_ext(gsub(".gz", "", grn_location))
    print(grn_location)
  if (extension %in% c("txt", "tsv")){
    X = read.table(grn_location, header = F, row.names = NULL, stringsAsFactors = F, ... )
  } else if (extension=="csv"){
    X = read.csv  (grn_location, header = F, row.names = NULL, stringsAsFactors = F, ... ) 
  } else {
    stop("Unknown format!")
  }
  
  # add score of -1 if missing
  if(ncol(X) == 2){
    X[[3]] = -1
  }
  
  # Fix meanings of columns
  first_column_meaning = load_grn_metadata()[grn_name, "first_column"]
  if(first_column_meaning %in% c("symmetric", EXPECTED_GRN_COLNAMES[[1]])){
    colnames(X) = EXPECTED_GRN_COLNAMES
  } else if(first_column_meaning == EXPECTED_GRN_COLNAMES[[2]]){
    colnames(X) = EXPECTED_GRN_COLNAMES[c(2, 1, 3)]
  } else {
    stop("Can't tell if target or regulator comes first! Fix the metadata.\n")
  }
  
  if(sort_by_weight){
    X = X[order(X[[EXPECTED_GRN_COLNAMES[[3]]]], decreasing = T),]
  }
  
  return(X[EXPECTED_GRN_COLNAMES])
}
```

```r
load_geneset = function( geneset_name, uppercase = T,  ... ){
    geneset_location = here(geneset_name )
  if(is.null(geneset_location)){
    stop("Error locating geneset! Please report this error to Eric.\n")
  }
  X = read.csv(geneset_location, header = F, row.names = NULL, ..., stringsAsFactors = F )[[1]]
  if(uppercase){
    X %<>% toupper
  }
  if(toupper(X[[1]] == "GENE")){
    X = X[-1]
  }
  return(X)
}
```
